```python
NYC_df['V1Sedan']=0
NYC_df['V1SWorSUV']=0

NYC_df.loc[NYC_df.V1 == 'Sedan', 'V1Sedan'] = 1
NYC_df.loc[NYC_df.V1 == '4 dr sedan', 'V1Sedan'] = 1
NYC_df.loc[NYC_df.V1 == '2 dr sedan', 'V1Sedan'] = 1
NYC_df.loc[NYC_df.V1 == 'Convertible', 'V1Sedan'] = 1

NYC_df.loc[NYC_df.V1 == 'SPORT UTILITY / STATION WAGON', 'V1SWorSUV'] = 1
NYC_df.loc[NYC_df.V1 == 'Station Wagon/Sport Utility Vehicle', 'V1SWorSUV'] = 1


NYC_df['V2Sedan']=0
NYC_df['V2SWorSUV']=0

NYC_df.loc[NYC_df.V2 == 'Sedan', 'V2Sedan'] = 1
NYC_df.loc[NYC_df.V2 == '4 dr sedan', 'V2Sedan'] = 1
NYC_df.loc[NYC_df.V2 == '2 dr sedan', 'V2Sedan'] = 1
NYC_df.loc[NYC_df.V2 == 'Convertible', 'V2Sedan'] = 1

NYC_df.loc[NYC_df.V2 == 'SPORT UTILITY / STATION WAGON', 'V2SWorSUV'] = 1
NYC_df.loc[NYC_df.V2 == 'Station Wagon/Sport Utility Vehicle', 'V2SWorSUV'] = 1


NYC_df['V1_sum']=NYC_df['V1Sedan']+NYC_df['V1SWorSUV']
NYC_df['V2_sum']=NYC_df['V2Sedan']+NYC_df['V2SWorSUV']


NYC_df.loc[NYC_df.V1_sum != 1, 'drop_flag'] = 1
NYC_df.loc[NYC_df.V2_sum != 1, 'drop_flag'] = 1

NYC_df = NYC_df[NYC_df["drop_flag"] == 0]



```

```python
NYC_df['MorningRush']=0
NYC_df['Midday']=0
NYC_df['EveningRush']=0
NYC_df['Overnight']=0

from datetime import datetime
NYC_df['TIME INT'] = NYC_df['TIME'].apply(lambda x: datetime.strptime(x, '%H:%M').hour)

NYC_df['MorningRush']= NYC_df['TIME INT'].apply(lambda x: int(x >= 6 and x < 10))
NYC_df['Midday']= NYC_df['TIME INT'].apply(lambda x: int(x >= 10 and x < 16))
NYC_df['EveningRush']= NYC_df['TIME INT'].apply(lambda x: int(x >= 16 and x < 20))
NYC_df['Overnight']= NYC_df['TIME INT'].apply(lambda x: int(x >= 20 or x < 6))

NYC_df['Time_set_sum']=NYC_df['MorningRush']+NYC_df['Midday']+NYC_df['EveningRush']+NYC_df['Overnight']
NYC_df.loc[NYC_df.Time_set_sum != 1, 'drop_flag'] = 1


NYC_df = NYC_df[NYC_df["drop_flag"] == 0]
```
