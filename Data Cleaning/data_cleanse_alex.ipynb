```python
# Dependencies and Setup
import pandas as pd
import numpy as np

headers=['DATE','TIME','BOROUGH','drop1','LATITUDE','LONGITUDE','drop2','drop3','drop4','drop5','drop6',
'drop7','drop8','drop9','drop10','drop11','NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED',
'CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2','CONTRIBUTING FACTOR VEHICLE 3','drop12',
'drop13','drop14','VEHICLE TYPE CODE 1','VEHICLE TYPE CODE 2','VEHICLE TYPE CODE 3',
'drop15','drop16']

dtypes={'DATE': str,'TIME': str,'BOROUGH': str,'drop1': str,'LATITUDE': float,'LONGITUDE':float,'drop2': str,'drop3': str,
'drop4': str,'drop5': str,'drop6': str,'drop7': str,'drop8': str,'drop9': str,'drop10': str,'drop11': str,
'NUMBER OF MOTORIST INJURED': int,'NUMBER OF MOTORIST KILLED':int,'CONTRIBUTING FACTOR VEHICLE 1': str,
'CONTRIBUTING FACTOR VEHICLE 2':str,'CONTRIBUTING FACTOR VEHICLE 3': str,'drop12': str,'drop13':str,'drop14': str,
'VEHICLE TYPE CODE 1':str,'VEHICLE TYPE CODE 2': str,'VEHICLE TYPE CODE 3':str,'drop15':str,'drop16': str}

parse_dates=['DATE']

NYC_df = pd.read_csv('NYPD_Motor_Vehicle_Collisions.csv', skiprows=[0],header=None, names=headers, dtype=dtypes, parse_dates=parse_dates)
```

```python
NYC_df.head()
```

```python
#drop the drop columns
NYC_df=NYC_df.drop(columns=['drop1','drop2','drop3','drop4','drop5','drop6','drop7','drop8','drop9','drop10','drop11','drop12','drop13','drop14','drop15','drop16'])
```

```python
#shorten the names of some columns
NYC_df = NYC_df.rename(columns={"LATITUDE": "LAT"})
NYC_df = NYC_df.rename(columns={"LONGITUDE": "LONG"})
NYC_df = NYC_df.rename(columns={"CONTRIBUTING FACTOR VEHICLE 1": "CFV1"})
NYC_df = NYC_df.rename(columns={"CONTRIBUTING FACTOR VEHICLE 2": "CFV2"})
NYC_df = NYC_df.rename(columns={"CONTRIBUTING FACTOR VEHICLE 3": "CFV3"})
NYC_df = NYC_df.rename(columns={"VEHICLE TYPE CODE 1": "V1"})
NYC_df = NYC_df.rename(columns={"VEHICLE TYPE CODE 2": "V2"})
NYC_df = NYC_df.rename(columns={"VEHICLE TYPE CODE 3": "V3"})
```

```python
NYC_df.head()
```

```python
NYC_df['Manhattan']=NYC_df.BOROUGH.apply(lambda x: int(x == 'MANHATTAN'))
NYC_df['Brooklyn']=NYC_df.BOROUGH.apply(lambda x: int(x == 'QUEENS'))
NYC_df['Bronx']=NYC_df.BOROUGH.apply(lambda x: int(x == 'BRONX'))
NYC_df['Queens']=NYC_df.BOROUGH.apply(lambda x: int(x == 'BROOKLYN'))
NYC_df['Staten Island']=NYC_df.BOROUGH.apply(lambda x: int(x == 'STATEN ISLAND'))

#NYC_df.loc[NYC_df.BOROUGH == 'MANHATTAN', 'Manhattan'] = 1
#NYC_df.loc[NYC_df.BOROUGH == 'QUEENS', 'Queens'] = 1
#NYC_df.loc[NYC_df.BOROUGH == 'BRONX', 'Bronx'] = 1
#NYC_df.loc[NYC_df.BOROUGH == 'BROOKLYN', 'Brooklyn'] = 1
#NYC_df.loc[NYC_df.BOROUGH == 'STATEN ISLAND', 'Staten Island'] = 1
NYC_df['Borough_sum']=NYC_df['Manhattan']+NYC_df['Queens']+NYC_df['Brooklyn']+NYC_df['Staten Island']+NYC_df['Bronx']
```

```python
#If any of the borough columns do not contain a 1 add a 1 to the drop column
#NYC_df.loc['Borough_sum' != 1, 'drop'] = 1
NYC_df['drop'] = NYC_df['Borough_sum'].apply( lambda x: int(x != 1))

NYC_df = NYC_df[NYC_df["drop"] == 0]
NYC_df.count()
```

```python
NYC_df.head()
```

```python
NYC_df['Driver Inattention/Distraction'] = NYC_df.CFV1.apply(lambda x:  int(x == 'Driver Inattention/Distraction'))
NYC_df['Failure to Yield Right-of-Way']=NYC_df.CFV1.apply(lambda x:  int(x == 'Failure to Yield Right-of-Way'))
NYC_df['Following Too Closely']=NYC_df.CFV1.apply(lambda x:  int(x == 'Following Too Closely'))
NYC_df['Backing Unsafely']=NYC_df.CFV1.apply(lambda x:  int(x == 'Backing Unsafely'))
NYC_df['Unsafe Lane Changing']=NYC_df.CFV1.apply(lambda x:  int(x == 'Unsafe Lane Changing'))
NYC_df['Traffic Control Disregarded']=NYC_df.CFV1.apply(lambda x:  int(x == 'Traffic Control Disregarded'))
NYC_df['Unsafe Speed']=NYC_df.CFV1.apply(lambda x:  int(x == 'Unsafe Speed'))
NYC_df['Turning Improperly']=NYC_df.CFV1.apply(lambda x:  int(x == 'Turning Improperly'))
NYC_df['Aggressive Driving/Road Rage']=NYC_df.CFV1.apply(lambda x:  int(x == 'Aggressive Driving/Road Rage'))
NYC_df['Failure to Keep Right']=NYC_df.CFV1.apply(lambda x:  int(x == 'Failure to Keep Right'))

#NYC_df.loc[NYC_df.CFV1 == 'Driver Inattention/Distraction', 'Driver Inattention/Distraction'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Failure to Yield Right-of-Way', 'Failure to Yield Right-of-Way'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Following Too Closely', 'Following Too Closely'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Backing Unsafely', 'Backing Unsafely'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Unsafe Lane Changing', 'Unsafe Lane Changing'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Traffic Control Disregarded', 'Traffic Control Disregarded'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Unsafe Speed', 'Unsafe Speed'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Turning Improperly', 'Turning Improperly'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Aggressive Driving/Road Rage', 'Aggressive Driving/Road Rage'] = 1
#NYC_df.loc[NYC_df.CFV1 == 'Failure to Keep Right', 'Failure to Keep Right'] = 1


NYC_df['CF_sum']=NYC_df['Driver Inattention/Distraction']+NYC_df['Failure to Yield Right-of-Way']+NYC_df['Following Too Closely']+NYC_df['Backing Unsafely']+NYC_df['Unsafe Lane Changing']+NYC_df['Traffic Control Disregarded']+NYC_df['Unsafe Speed']+NYC_df['Turning Improperly']+NYC_df['Aggressive Driving/Road Rage']+NYC_df['Failure to Keep Right']
```

```python
#If any of the Contributing Factor V1 columns do not contain a 1 add a 1 to the drop column
#NYC_df.loc['CF_sum' != 1, 'drop'] = 1

NYC_df['drop'] = NYC_df['CF_sum'].apply( lambda x: int(x != 1))

NYC_df = NYC_df[NYC_df["drop"] == 0]
NYC_df.count()
```

```python
#find and replace 'Station Wagon/Sport Utility Vehicle' with 'SPORT UTILITY / STATION WAGON' to combine catagories
NYC_df = NYC_df.replace({'Station Wagon/Sport Utility Vehicle': 'SPORT UTILITY / STATION WAGON'}, regex=True)
```

```python
#find and replace 'Sedan' with 'PASSENGER VEHICLE' to combine catagories
NYC_df = NYC_df.replace({'Sedan': 'PASSENGER VEHICLE'}, regex=True)
```

```python
NYC_df['V1 PASSENGER VEHICLE'] = NYC_df.V1.apply(lambda x: int(x == 'PASSENGER VEHICLE'))
NYC_df['V1 SPORT UTILITY / STATION WAGON'] = NYC_df.V1.apply(lambda x : int(x == 'SPORT UTILITY / STATION WAGON'))

#NYC_df.loc[, 'V1 PASSENGER VEHICLE'] = 1
#NYC_df.loc[, 'V1 SPORT UTILITY / STATION WAGON'] = 1
NYC_df['V1_sum']=NYC_df['V1 PASSENGER VEHICLE']+NYC_df['V1 SPORT UTILITY / STATION WAGON']
```

```python
#If any of the V1 columns do not contain a 1 add a 1 to the drop column
NYC_df['drop'] = NYC_df['V1_sum'].apply( lambda x: int(x != 1))

NYC_df = NYC_df[NYC_df["drop"] == 0]

NYC_df.count()
```

```python
NYC_df['V2 PASSENGER VEHICLE'] = NYC_df.V2.apply(lambda x: int(x == 'PASSENGER VEHICLE'))
NYC_df['V2 SPORT UTILITY / STATION WAGON'] = NYC_df.V2.apply(lambda x: int(x == 'SPORT UTILITY / STATION WAGON'))

#NYC_df.loc[NYC_df.V1 == 'PASSENGER VEHICLE', 'V2 PASSENGER VEHICLE'] = 1
#NYC_df.loc[NYC_df.V1 == 'SPORT UTILITY / STATION WAGON', 'V2 SPORT UTILITY / STATION WAGON'] = 1

NYC_df['V2_sum']=NYC_df['V2 PASSENGER VEHICLE']+NYC_df['V2 SPORT UTILITY / STATION WAGON']

#If any of the V2 columns do not contain a 1 add a 1 to the drop column
#NYC_df.loc['V1_sum' != 1, 'drop'] = 1

#NYC_df.head()

NYC_df['drop'] = NYC_df['V2_sum'].apply(lambda x: int(x != 1))

NYC_df = NYC_df[NYC_df["drop"] == 0]

NYC_df.count()
```

```python
NYC_df['MOTORIST INJURED']=NYC_df['NUMBER OF MOTORIST INJURED'].apply(lambda x: int(x>0))
NYC_df['MOTORIST KILLED']=NYC_df['NUMBER OF MOTORIST KILLED'].apply(lambda x: int(x>0))
#NYC_df['MOTORIST NOT INJURED']=0
#NYC_df['MOTORIST NOT KILLED']=0

NYC_df['MOTORIST HARMED'] = NYC_df['NUMBER OF MOTORIST INJURED'] + NYC_df['NUMBER OF MOTORIST KILLED'] > 0
NYC_df['MOTORIST HARMED'] = NYC_df['MOTORIST HARMED'].apply(lambda x: int(x))

NYC_df.count()

#NYC_df.loc[NYC_df['NUMBER OF MOTORIST INJURED'] >0, 'MOTORIST INJURED'] = 1
#NYC_df.loc[NYC_df['NUMBER OF MOTORIST KILLED'] >0, 'MOTORIST KILLED'] = 1
#NYC_df.loc[NYC_df['NUMBER OF MOTORIST INJURED'] + NYC_df['NUMBER OF MOTORIST KILLED'] ==0, 'MOTORIST NOT INJURED OR KILLED'] = 1
#NYC_df['INJURED_KILLED_sum']=NYC_df['MOTORIST INJURED']+NYC_df['MOTORIST KILLED']
#If any of the V2 columns do not contain a 1 add a 1 to the drop column
# NYC_df.loc['INJURED_KILLED_sum' != 1, 'drop'] = 1
#NYC_df.head()
```

```python
NYC_df = NYC_df.drop(columns = ["BOROUGH","LAT","LONG","CFV1","CFV2","CFV3","NUMBER OF MOTORIST INJURED","NUMBER OF MOTORIST KILLED","V1","V2","V3"])
```
