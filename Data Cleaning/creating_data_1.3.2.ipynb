```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "data_1.2.csv"

# Read file and store into pandas data frame
df = pd.read_csv(file_to_load)

#print intial summary head() table
df.head()
```

```python
#droping the columns listed below:
df = df.drop(columns=['LOCATION', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED', 'NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED', 'NUMBER OF MOTORIST KILLED'])
```

```python
df.head()
```

```python
df.count()
```

```python
#Drop all rows not containing a Vehicle One
df = df.dropna(subset = ['V1'])
```

```python
df.count()
```

```python
#Drop all rows not containing a Contributing Factor for Vehicle One
df = df.dropna(subset = ['CFV1'])
```

```python
df.count()
```

```python
df['V1'].value_counts()
```

```python
#filtering out all vehicle one types beside those presented below:
v1_vehicle_types = ["PASSENGER VEHICLE", "SPORT UTILITY / STATION WAGON", "Sedan", "Station Wagon/Sport Utility Vehicle"]
df = df.loc[df['V1'].isin(v1_vehicle_types)]
```

```python
df["V1"].value_counts()
```

```python
# Leaves us with 183,826 vehicle one contributing factors, 129,555 vehicle two contributing factors, 
# and 54,271 cases where vehicle one was the only party with a contributing factor listed
df.count()
```

```python
df.to_csv('data_1.3.2.csv', index=False)
```
