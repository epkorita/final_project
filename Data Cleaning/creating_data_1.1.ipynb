```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "data_1.0.csv"

# Read purchasing file and store into pandas data frame
df = pd.read_csv(file_to_load)

#print intial summary head() table
df.head()
```

```python
df.dropna(subset = ['BOROUGH', 'LATITUDE', 'LONGITUDE'])
```

```python
df.head()
```

```python
df = df.drop(columns=['ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME'])
```

```python
df.head()
```

```python
df = df.drop(columns=['UNIQUE KEY', 'CONTRIBUTING FACTOR VEHICLE 3', 'CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5'])
```

```python
df.head()
```

```python
df = df.drop(columns=['ZIP CODE'])
```

```python
df.head()
```

```python
df = df.rename(columns={"LATITUDE": "LAT"})
df = df.rename(columns={"LONGITUDE": "LONG"})
```

```python
df.head()
```

```python
df = df.rename(columns={"CONTRIBUTING FACTOR VEHICLE 1": "CFV1"})
df = df.rename(columns={"CONTRIBUTING FACTOR VEHICLE 2": "CFV2"})
```

```python
df.head()
```

```python
#Drop rows that contain Unspecified in CFV1
df = df[df.CFV1 != 'Unspecified']
```

```python
df.head()
```

```python
#Drop rows that contain Unspecified in CFV2
df = df[df.CFV2 != 'Unspecified']
```

```python
df.head()
```

```python
df.dropna(subset = ['CFV1', 'CFV2', 'BOROUGH'])
```

```python
df.to_csv('data_1.1.csv', index=False)
```
